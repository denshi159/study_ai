```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import confusion_matrix
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
```

sys.pathの設定

```python
cancer_df = pd.read_csv('/content/drive/My Drive/study_ai_ml_google/data/cancer.csv')
```

```python
print('cancer df shape: {}'.format(cancer_df.shape))
```

```python
cancer_df
```

```python
cancer_df.drop('Unnamed: 32', axis=1, inplace=True)
cancer_df
```

・diagnosis:診断結果（良性がB／悪性がM）・説明変数は３列以降、目的変数を２列目としロジスティック回帰で分類

```python
#目的変数の抽出
y = cancer_df.diagnosis.apply(lambda d: 1 if d == 'M' else 0)
```

```python
#説明変数の抽出
X = cancer_df.loc[:, 'radius_mean':]
```

```python
#学習用とテスト用でデータを分離
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#標準化
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#ロジスティック回帰で学習
logistic = LogisticRegressionCV(cv=10, random_state=0)
logistic.fit(X_train_scaled, y_train)

#検証
print('Train score: {:.3f}'.format(logistic.score(X_train_scaled, y_train)))
print('Test score: {:.3f}'.format(logistic.score(X_test_scaled, y_test)))
print('Consusion matrix:\n{}'.format(confusion_matrix(y_true=y_test, y_pred=logistic.predict(X_test_scaled))))

```

```python
#説明変数を変更
Z = cancer_df.loc[:,'radius_mean':'area_mean']
```

```python
#学習用とテスト用でデータを分離
Z_train, Z_test, y_train, y_test = train_test_split(Z, y, random_state=0)

#標準化
scaler = StandardScaler()
Z_train_scaled = scaler.fit_transform(Z_train)
Z_test_scaled = scaler.transform(Z_test)

#ロジスティック回帰で学習
logistic = LogisticRegressionCV(cv=10, random_state=0)
logistic.fit(Z_train_scaled, y_train)

#検証
print('Train score: {:.3f}'.format(logistic.score(Z_train_scaled, y_train)))
print('Test score: {:.3f}'.format(logistic.score(Z_test_scaled, y_test)))
print('Consusion matrix:\n{}'.format(confusion_matrix(y_true=y_test, y_pred=logistic.predict(Z_test_scaled))))

```
